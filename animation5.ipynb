```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation, rc
from IPython.display import HTML

def init():
    ax.set_xlim(xmin, xmax)
    ax.set_ylim(ymin, ymax)
    ax.grid()
    return

def data_gen():
    cnt = 0
    t = 0.0
    dt = (xmax - xmin) / (nframes - 1)
    while cnt < nframes:
        yield t, np.sin(2*np.pi*t), 0.5 * np.cos(2*np.pi*t), 0.2 * np.cos(2*np.pi*t)
        cnt += 1
        t += dt

def run(data):
    t, y1, y2, y3 = data
    xdata.append(t)
    ydata1.append(y1)
    ydata2.append(y2)
    ydata3.append(y3)
    line1.set_data(xdata, ydata1)
    line2.set_data(xdata, ydata2)
    line3.set_data(xdata, ydata3)
    return line1, line2, line3

nframes =  30
xmin, xmax = (0.0, 3.0)
ymin, ymax = (-1.3, 1.3)

fig, ax = plt.subplots()
plt.close()

xdata, ydata1, ydata2, ydata3 = [], [], [], []
line1, = ax.plot(xdata, ydata1, lw=6, color='coral')
line2, = ax.plot(xdata, ydata2, lw=2, color='blue', linestyle='-.')
line3, = ax.plot(xdata, ydata2, lw=1, color='red', linestyle='-') 

anim = animation.FuncAnimation(fig, run, data_gen, init_func=init,
                              interval=50, repeat=False, blit=False,
                              save_count=nframes)
rc('animation', html='jshtml')
anim
```
