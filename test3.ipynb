```
import matplotlib.pyplot as plt
import numpy as np
```

```
x = np.arange(0.0, 1.0, 0.01)
y = np.tan(2.0 * np.pi * x)
```

```
plt.plot(x, y)
```

```
plt.ylim(-1.5, 1.5)
plt.grid()
plt.plot(x, y)
```

```
plt.grid()
plt.plot(x, np.sin(2.0 * np.pi * x))
plt.plot(x, np.cos(2.0 * np.pi * x))
```
